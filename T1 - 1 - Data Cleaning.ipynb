# Carga de datos a través de la funcion OPEN
## Todo esto se hace por que estoy leyendo un archivo de texto, por lo tanto es un STRING

```python
filename3 = "Customer Churn Model.txt" #importamos el csv donde estan los nombres de las columnas
df3 = open(filepath + filename3,"r")
```

```python
cols = df3.readline().strip().split(",") #strip elimina todo lo blanco delante y detras
cols
```

```python
n_cols = len(cols)
n_cols
```

```python
counter = 0
main_dict ={}
for col in cols:
    main_dict[col] = []
```

```python
for line in df3:
    valores = line.strip().split(",")
    for i in range(len(cols)):
       main_dict[cols[i]].append(valores[i])
    counter += 1

print("El data set tiene %d filas y %d columas"%(counter,n_cols)) #me tendria q dar 3333 filas y 21 columnas
```
