# Agregacion de datos por categoría

```python
import numpy as np
import pandas as pd
```

```python
gender = ["Male", "Female"]
income = ["Poor", "Middle Class", "Rich"]
```

```python
n = 500
gender_data = []
income_data = []

for i in range(0, n):
    gender_data.append(np.random.choice(gender))
    income_data.append(np.random.choice(income))

gender_data[1:10]

```

```python
income_data[1:10]
```

```python
#Z - > N(0,1)
#N(m,s) -> m + s * Z
height = 160 + 30 * np.random.randn(n)
weight = 65 + 25 * np.random.randn(n)
age = 30 + 12 * np.random.randn(n)
income = 18000 + 3500 * np.random.randn(n)
```

```python
data = pd.DataFrame(
    {
        "Gender" : gender_data,
        "Economic Status" : income_data,
        "Height" : height,
        "Weight" : weight,
        "Age" : age,
        "Income" : income
    }
)
```

```python
data
```

## Agrupación de datos

```python
data.groupby("Gender")
```

```python
# crea un objeto de nombre groupby
```

```python
grouped_gender = data.groupby("Gender")
```

```python
grouped_gender.groups
```

```python
grouped_gender.get_group("Female")
```

```python
# Agrupacion por mas de una categoría
```

```python
double_gropu = data.groupby(["Gender", "Economic Status"])
```

```python
double_gropu.groups
```

```python
for names, groups in double_gropu:
    print(names)
    print(groups)
```

## Agregacion de datos
* Operaciones sobre datos agrupados

```python
double_gropu.sum()
```

```python
double_gropu.mean()
```

```python
double_gropu.size()
```

```python
double_gropu.describe() # sale mucha info
```

```python
grouped_income = double_gropu["Income"] # me quedo con una sola de las variables
grouped_income.describe()
```

```python
double_gropu.aggregate(
    {
        "Income" : np.sum,
        "Age": np.mean,
        "Height" : np.std
        }
)
```

```python
double_gropu.aggregate(
    {
        "Age": np.mean,
        "Height" : lambda h:(np.mean(h)/np.std(h)) # por is quiero hacer alguna funcion
        }
)
```

```python
double_gropu.aggregate([np.sum,np.mean,np.std]) # como ya esta doble filtrado por gender y economic status, aplica las funciones suma, media y desviacion standar a las columnas restantes
```

## Filtrado de datos

```python
double_gropu.sum()
```

```python
#quiero quedarme con los grupos que la suma sea mayo a X
double_gropu["Age"].filter(lambda x: x.sum()>2400)
```

## Transofmacion de variables

```python
zscore = lambda x : (x - x.mean()/x.std())
```

```python
zgroup = double_gropu.transform(zscore)
zgroup
```

```python
import matplotlib.pyplot as plt
```

```python
plt.hist(zgroup["Age"])
```

```python
fill_na_mean = lambda x : x.fillna(x.mean())
```

```python
double_gropu.transform(fill_na_mean) #podriamos reemplazar los NA con el promedio de los valores por ejemplo
```
